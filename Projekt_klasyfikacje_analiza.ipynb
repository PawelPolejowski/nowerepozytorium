```python
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df1 = pd.read_csv('df1.csv')
df2 = pd.read_csv('df2.csv')
```

```python
merged_df = pd.merge(df1, df2, on="EmployeeNumber")
merged_df = merged_df.drop(merged_df[merged_df['Age'] > 60].index)
merged_df = merged_df.drop('Over18', axis=1)
merged_df.set_index("EmployeeNumber", inplace=True)
merged_df.sort_index(inplace=True)

merged_df

merged_df.head()
```

```python
merged_df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select columns with categorical data
categorical_columns = merged_df.select_dtypes(include=['object']).columns

# Create subplots
fig, ax = plt.subplots(2, 4, figsize=(20, 10))

# Iterate through categorical columns and plot count for each
for column, subplot in zip(categorical_columns, ax.flatten()):
    category_counts = merged_df[column].value_counts().sort_values(ascending=False)
    sns.countplot(data=merged_df, x=column, order=category_counts.index, ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(45)  # Rotate labels by 45 degrees

# Set common title for the subplots
fig.suptitle('Count of Employees by Categorical Variables')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()

```

```python
numerical = merged_df.select_dtypes(include='number').columns.tolist()
plt.subplots_adjust(hspace=4, wspace=0.3)

merged_df[numerical].hist(bins=15, figsize=(20, 20), layout=(7, 4))
```

```python
# Zakadajc, e masz DataFrame o nazwie 'merged_df'
merged_df["Attrition"] = merged_df["Attrition"].replace({"No": 0, "Yes": 1})

# Obliczenie macierzy korelacji
correlation_matrix = merged_df.corr()

# Wygenerowanie wykresu korelacji
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Wykres korelacji')
plt.show()
```
